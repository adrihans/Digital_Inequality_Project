## Small test

We can show and compare the temporal evolution of the number of tweets/retweets... and Factiva and Arcom for each candidate. 


We can then have a visiual representation of a possible correlation. 

### Arcom

```python
#Changing the way dates are handled in the tweets df:


new_df_tweets = pd.DataFrame()

for cand in names_candidates:
    df_cand = pd.DataFrame()
    
    
    for start_date, end_date in zip(list(df_arcom[df_arcom['candidat'] == cand].sort_values('start_date')['start_date']), 
                                    df_arcom[df_arcom['candidat'] == cand].sort_values('start_date')['end_date']):
    
        
        df_temp_dates = df_tweets_10RT[(df_tweets_10RT['candidate'] == cand) & 
                                (df_tweets_10RT['day'] >= start_date) & 
                                (df_tweets_10RT['day'] < end_date)].copy()
        
        try:
            dict_data = {'candidate' : [df_temp_dates['candidate'].unique()[0]],
                         'start_date' : [start_date], 
                         'end_date' : [end_date],
                        'tweet_count': [df_temp_dates['tweet_count'].sum()], 
                         'retweet_count' : [df_temp_dates['retweet_count'].sum()], 
                         'like_count':[df_temp_dates['like_count'].sum()],
                       'reply_count':[df_temp_dates['reply_count'].sum()]}


            df_ = pd.DataFrame(data = dict_data)
            df_cand = pd.concat([df_cand, df_])
        
        except:
            pass
        
        
    

    #Calculating percent evolution: 
    df_cand = df_cand.sort_values('start_date')
    
    for col in ['tweet_count', 'retweet_count', 'like_count', 'reply_count']:
        
        df_cand[col+'_pct_change'] = df_cand[col].pct_change()
        
    
    new_df_tweets = pd.concat([new_df_tweets, df_cand])
```

```python
new_df_tweets.shape
```

```python
new_df_tweets.head()
```

Calculating pct_change for arcom too:

```python
new_df_arcom = pd.DataFrame()

for cand in names_candidates:
    
    df_temp = df_arcom[df_arcom['candidat']==cand]
    df_temp = df_temp.sort_values('start_date')
    
    df_temp['time_diff_pct_change'] = df_temp['time_diff'].pct_change()
    
    new_df_arcom = pd.concat([new_df_arcom, df_temp])
```

Expending the dates:

```python
#Factiva:
new_df_arcom['date_new'] = new_df_arcom.apply(lambda x: pd.date_range(start=x['start_date'], end=x['end_date']), axis=1)
new_df_arcom = new_df_arcom.explode('date_new').reset_index(drop=True)

#Tweets:
new_df_tweets['date_new'] = new_df_tweets.apply(lambda x: pd.date_range(start=x['start_date'], end=x['end_date']), axis=1)
new_df_tweets = new_df_tweets.explode('date_new').reset_index(drop=True)
```

```python
for candidate in names_candidates:
    fig, ax1 = plt.subplots(figsize=(15,5))

    plt.title(candidate)

    x_arcom = np.array(new_df_arcom[new_df_arcom['candidat']==candidate].sort_values('date_new')['date_new'])
    y_arcom = np.array(new_df_arcom[new_df_arcom['candidat']==candidate].sort_values('date_new')['time_diff_pct_change'])

    x_tw = np.array(new_df_tweets[new_df_tweets['candidate']==candidate].sort_values('date_new')['date_new'])
    y_tw = np.array(new_df_tweets[new_df_tweets['candidate']==candidate].sort_values('date_new')['retweet_count_pct_change'])



    ax1.plot(x_arcom, y_arcom, color='b')
    ax1.set_ylabel('relative evolution', color='b')

    ax2 = ax1.twinx()
    ax2.set_ylabel('tweets', color='r')
    
    ax1.plot(x_tw, y_tw, color='r')
    plt.show()
```

### Factiva

```python
#Changing the way dates are handled in the tweets df:


new_df_tweets = pd.DataFrame()

for cand in names_candidates:
    df_cand = pd.DataFrame()
    
    
    for start_date, end_date in zip(list(df_factiva[df_factiva['candidat'] == cand].sort_values('start_date')['start_date']), 
                                    df_factiva[df_factiva['candidat'] == cand].sort_values('start_date')['end_date']):
    
        
        df_temp_dates = df_tweets_10RT[(df_tweets_10RT['candidate'] == cand) & 
                                (df_tweets_10RT['day']>= start_date) & 
                                (df_tweets_10RT['day']<=end_date)].copy()
        
        try:
            dict_data = {'candidate' : [df_temp_dates['candidate'].unique()[0]],
                         'start_date' : [start_date], 
                         'end_date' : [end_date],
                        'tweet_count': [df_temp_dates['tweet_count'].sum()], 
                         'retweet_count' : [df_temp_dates['retweet_count'].sum()], 
                         'like_count':[df_temp_dates['like_count'].sum()],
                       'reply_count':[df_temp_dates['reply_count'].sum()]}


            df_ = pd.DataFrame(data = dict_data)
            df_cand = pd.concat([df_cand, df_])
        
        except:
            pass
        
        
    

    #Calculating percent evolution: 
    df_cand = df_cand.sort_values('start_date')
    
    for col in ['tweet_count', 'retweet_count', 'like_count', 'reply_count']:
        
        df_cand[col+'_pct_change'] = df_cand[col].pct_change()
        
    
    new_df_tweets = pd.concat([new_df_tweets, df_cand])
```

```python
new_df_tweets.shape
```

```python
new_df_tweets.head()
```

Calculating pct_change for factiva too:

```python
new_df_factiva = pd.DataFrame()

for cand in names_candidates:
    
    df_temp = df_factiva[df_factiva['candidat']==cand]
    df_temp = df_temp.sort_values('start_date')
    
    df_temp['Nombres de documents_pct_change'] = df_temp['Nombres de documents'].pct_change()
    
    new_df_factiva = pd.concat([new_df_factiva, df_temp])
```

Expending the dates:

```python
#Factiva:
new_df_factiva['date_new'] = new_df_factiva.apply(lambda x: pd.date_range(start=x['start_date'], end=x['end_date']), axis=1)
new_df_factiva = new_df_factiva.explode('date_new').reset_index(drop=True)

#Tweets:
new_df_tweets['date_new'] = new_df_tweets.apply(lambda x: pd.date_range(start=x['start_date'], end=x['end_date']), axis=1)
new_df_tweets = new_df_tweets.explode('date_new').reset_index(drop=True)
```

```python
for candidate in names_candidates:
    fig, ax1 = plt.subplots()

    plt.title(candidate)

    x_factiva = np.array(new_df_factiva[new_df_factiva['candidat']==candidate].sort_values('date_new')['date_new'])
    y_factiva = np.array(new_df_factiva[new_df_factiva['candidat']==candidate].sort_values('date_new')['Nombres de documents_pct_change'])

    x_tw = np.array(new_df_tweets[new_df_tweets['candidate']==candidate].sort_values('date_new')['date_new'])
    y_tw = np.array(new_df_tweets[new_df_tweets['candidate']==candidate].sort_values('date_new')['retweet_count_pct_change'])



    ax1.plot(x_factiva, y_factiva, color='b')
    ax1.set_ylabel('relative evolution', color='b')

    ax2 = ax1.twinx()
    ax2.set_ylabel('tweets', color='r')
    
    ax1.plot(x_tw, y_tw, color='r')
    plt.show()
```

### Both?

```python
df_arc_fac = pd.merge(left = new_df_factiva, 
         right = new_df_arcom, 
         left_on = ['date_new', 'candidat'],
         right_on = ['date_new', 'candidat'])[['date_new', 'candidat', 
                                 'Nombres de documents_pct_change', 
                                 'time_diff_pct_change']].rename(columns = {"candidat_x":'candidat'})


df_arc_fac_tw = pd.merge(left = df_arc_fac, 
                        right = new_df_tweets, 
                        left_on = ['date_new', 'candidat'],
                        right_on = ['date_new', 'candidate'])[['date_new', 'candidat', 
                                 'Nombres de documents_pct_change', 
                                 'time_diff_pct_change', 'retweet_count_pct_change']]      
```

```python
df_arc_fac_tw.dropna(inplace = True)
df_arc_fac_tw['mean_media_pct_change'] = df_arc_fac_tw[['Nombres de documents_pct_change', 
                                                        'time_diff_pct_change']].mean(axis=1)
```

#### Cor

```python
df_arc_fac_tw.corr()
```

```python
for candidate in names_candidates:
    df_temp = df_arc_fac_tw[df_arc_fac_tw['candidat']==candidate]
    print(candidate)
    print('corr RT & time : {:.2f}'.format(df_temp['retweet_count_pct_change'].corr(df_temp['time_diff_pct_change'])))
    print('corr RT & factiva : {:.2f}'.format(df_temp['retweet_count_pct_change'].corr(df_temp['Nombres de documents_pct_change'])))
    
    print('corr RT & mean : {:.2f}'.format(df_temp['retweet_count_pct_change'].corr(df_temp['mean_media_pct_change'])))
```

#### Viz

```python
for candidate in names_candidates:
    fig, ax1 = plt.subplots()

    plt.title(candidate)

    x = np.array(df_arc_fac_tw[df_arc_fac_tw['candidat']==candidate].sort_values('date_new')['date_new'])
    y_media = np.array(df_arc_fac_tw[df_arc_fac_tw['candidat']==candidate].sort_values('date_new')['mean_media_pct_change'])
    y_tw = np.array(df_arc_fac_tw[df_arc_fac_tw['candidat']==candidate].sort_values('date_new')['retweet_count_pct_change'])

    

    ax1.plot(x, y_media, color='b')
    ax1.set_ylabel('relative evolution', color='b')

    ax2 = ax1.twinx()
    ax2.set_ylabel('tweets', color='r')
    
    ax1.plot(x, y_tw, color='r')
    plt.show()
```
