```python
#Changing the way dates are handled in the tweets df:


new_df_tweets = pd.DataFrame()

for cand in names_candidates:
    df_cand = pd.DataFrame()
    
    
    for start_date, end_date in zip(list(df_arcom[df_arcom['candidat'] == cand].sort_values('start_date')['start_date']), 
                                    df_arcom[df_arcom['candidat'] == cand].sort_values('start_date')['end_date']):
    
        
        df_temp_dates = df_tweets_10RT[(df_tweets_10RT['candidate'] == cand) & 
                                (df_tweets_10RT['day'] >= start_date) & 
                                (df_tweets_10RT['day'] < end_date)].copy()
        
        dict_data = {'candidate' : [df_temp_dates['candidate'].unique()[0]],
                     'start_date' : [start_date], 
                     'end_date' : [end_date],
                    'tweet_count': [df_temp_dates['tweet_count'].sum()], 
                     'retweet_count' : [df_temp_dates['retweet_count'].sum()], 
                     'like_count':[df_temp_dates['like_count'].sum()],
                   'reply_count':[df_temp_dates['reply_count'].sum()]}

        df_ = pd.DataFrame(data = dict_data)
        df_cand = pd.concat([df_cand, df_])

    #Calculating percent evolution: 
    df_cand = df_cand.sort_values('start_date')
    
    for col in count_cols_twitter:
        
        df_cand[col+'_pct_change'] = df_cand[col].pct_change()
        
    
    new_df_tweets = pd.concat([new_df_tweets, df_cand])
```

```python
df_arcom_tw_copy = df_arcom_tw.drop(columns= 'retweet_count_pct_change').copy()
df_factiva_arcom_tw = pd.merge(left = df_factiva_tw, 
                              right = df_arcom_tw_copy, 
                              left_on = ['date_new', 'candidat'], 
                              right_on = ['date_new', 'candidat'])

#Dropping nans:
df_factiva_arcom_tw.dropna(inplace=True)
#Renaming some columns:
df_factiva_arcom_tw.rename(columns= {'retweet_count_pct_change':'RT', 
                                         'Nombres de documents_pct_change':'factiva', 
                                        'time_diff_pct_change' : 'arcom'}, inplace=True)
```

### Computing the regression analysis:

```python
#Creating a dataframe of coefficient for each type of media:
df_coef_media = pd.DataFrame()

for candidate in names_candidates:
    #Slicing the df on the candidate we're looping on:
    df_factiva_arcom_tw_temp = df_factiva_arcom_tw[df_factiva_arcom_tw['candidat'] == candidate].copy()
    #Retrieving the regression results for that candidate:
    result = sm.ols(formula="RT ~ factiva + arcom", data=df_factiva_arcom_tw_temp).fit()    
    
    
    #-------------------Saving the an image of the regression report------------------------
    #Source: https://stackoverflow.com/questions/46664082/python-how-to-save-statsmodels-results-as-image-file
    plt.rc('figure', figsize=(12, 7))
    #plt.text(0.01, 0.05, str(model.summary()), {'fontsize': 12}) old approach
    plt.text(0.01, 0.05, str(result.summary()), {'fontsize': 10}, fontproperties = 'monospace') # approach improved by OP -> monospace!
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('images/OLS_regression_summary/'+ candidate +'.png')
    plt.show()
    
    
    #------------------------Concat the DF of Coefficients:
    x = {'candidat' : [candidate]}
    y = {k : [result.params[k]] for k in result.params.keys()}
    df_coef_media = pd.concat([df_coef_media, pd.DataFrame(data = {**x, **y})])
```

### Viz considering the ponderated importance of each type?????

```python
def compute_w_avg(row):
    candidate = row['candidat']
    
    #------------Factiva data & weight-----------------
    factiva = row['factiva']
    w_factiva = df_coef_media[df_coef_media['candidat'] == candidate]['factiva']
    #------------Arcom data & weight-------------------
    arcom = row['arcom']
    w_arcom = df_coef_media[df_coef_media['candidat'] == candidate]['arcom']
    
    return (w_arcom*arcom + w_factiva*factiva) / (w_arcom+w_factiva)
#Computing the function:
df_factiva_arcom_tw['media_weighted_avg'] = df_factiva_arcom_tw.apply(compute_w_avg, axis=1)
```
